## Functions vs Methods

A **function** is a reusable block of code you call like `func(x)`. They do not belong to any object.

```python
# Function (built-in): operates on input
values = [" 10 ", " 20 ", " 30 "]
clean = list(map(str.strip, values))  # map() is a function; str.strip is a method used via function
print(clean)  # ['10', '20', '30']
```

A **method** is a function attached to an object, called like `obj.method()` and typically operates on that object's state. Are functions defined inside a specific class.


```python
# Method: attached to an object (string)
s = "  Data  "
print(s.strip())  # 'Data'
```

- Methods are functions tied to the class it is defined in, so it only exists for objects of that class (or subclasses).
- Functions are defined independently of any class and can operate on any objects that are passed to them. The key word is “class property.”
